```python
def convert_to_float_list(input_str):
    # Split the input string into a list of values
    values = input_str.split()
    
    # Convert each value to float and store in a list
    result_list = [float(value) for value in values]
    
    return result_list

# Example usage:
input_values = input("Enter space-separated values: ")
output_list = convert_to_float_list(input_values)

print("Result as a list of floats:", output_list)
```
